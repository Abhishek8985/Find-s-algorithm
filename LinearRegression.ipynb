```python
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
data=pd.read_csv("/content/headbrain.csv")
df=pd.DataFrame(data)
df=df.head(10)
df
```

```python
x=df[['Head Size(cm^3)']]
y=df[['Brain Weight(grams)']]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=3)
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
```

```python
y_test
```

```python
model.fit(x_train,y_train)
```

```python
pred=model.predict(x_test)
pred
```

```python
from sklearn.metrics import mean_squared_error
```

 In regression problems, you don't use accuracy as a metric because accuracy is typically used for classification problems. Instead, you should use regression evaluation metrics, such as Mean Squared Error (MSE), R-squared (R^2), or other relevant metrics.


```python
score=mean_squared_error(pred,y_test)
score
```
