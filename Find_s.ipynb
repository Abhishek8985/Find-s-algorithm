```python
import pandas as pd
```

```python
data=pd.read_csv("/content/ENJOYSPORT.csv")
df=pd.DataFrame(data)
df
```

```python
target=df.iloc[:,-1]
target
```

```python
concept=df.iloc[:,:-1]
concept
```

```python
def find_s(target,concept):
  h=list(concept.iloc[0])
  for i in range(len(target)):
    if target[i]== 1:
      for j in range(len(concept.iloc[i])):
                if concept.iloc[i, j] != h[j]:
                    h[j] = '?'
  print(h)
find_s(target,concept)
```
