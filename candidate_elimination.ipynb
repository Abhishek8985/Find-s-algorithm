```python
import pandas as pd
```

```python
data=pd.read_csv("/content/ENJOYSPORT.csv")
df=pd.DataFrame(data)
df
```

```python
target=df.iloc[:,-1]
target
```

```python
concept=df.iloc[:,:-1]
concept
```

```python
def candidate_elimination(target,concept):
  s=list(concept.iloc[0])
  g=[['?' for _ in range(len(s))] for _ in range(len(s))]
  for i in range(len(concept)):
    if target[i]==1:
      for j in range(len(s)):
        if concept.iloc[i][j] != s[j]:
          s[j]='?'
          g[j][j]='?'
    else:
      for j in range(len(s)):
        if s[j]!='?' and concept.iloc[i][j]!=s[j] :
          print(concept.iloc[i][j],s[j])
          g[j][j]=s[j]
  print("Specific hypo :",s)
  print("General hypo :",g)
candidate_elimination(target,concept)
```
